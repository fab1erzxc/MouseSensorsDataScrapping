```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from bs4 import BeautifulSoup as bs
import requests
```

```python
url = r'https://www.realgear.net/best-mouse-sensor-for-gaming/'

response = requests.get(url)

soup = bs(response.text, 'html.parser')
```

```python
soup
```

```python
content = soup.find_all('tr')
content
```

```python
content = content[0]
content1 = content.find_all('td')
content1 = [i.text.strip() for i in content1]
content1
```

```python
df = pd.DataFrame(columns = content1)
df
```

```python
content = soup.find_all('tr')
content = content[1:]
content = [i.find_all('td') for i in content]
content = [[i.text.strip() for i in j] for j in content]
content
```

```python
sensor_models = [i[0] for i in content]
sensor_models
```

```python
df
```

```python
df['Sensor Model'] = sensor_models
df
```

```python
manufacterer = [i[1] for i in content]
manufacterer
```

```python
df['Manufacturer'] = manufacterer
df
```

```python
df.drop('Manufacterer', axis=1, inplace=True)
```

```python
df
```

```python
sensors = [i[2] for i in content]
sensors
```

```python
df['Sensor'] = sensors
df
```

```python
flawless = [i[3] for i in content]
flawless
```

```python
df['Flawless'] = flawless
df
```

```python
content = content[:40]
```

```python
sensor_models = [i[0] for i in content]
df['Sensor Model'] = sensor_models
df
```

```python
df['Manufacturer'] = [i[1] for i in content]
df['Sensor'] = [i[2] for i in content]
df['Flawless'] = [i[3] for i in content]
df['DPI'] = [i[4] for i in content]
df['Tracking Speed'] = [i[5] for i in content]
df['Acceleration'] = [i[6] for i in content]
df 
```

```python
df.to_csv('sensor_data.csv', index=False)
```
